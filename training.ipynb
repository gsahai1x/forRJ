```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
plt.style.use('ggplot')
pd.options.display.float_format = '{:,.2f}'.format
from sklearn import set_config
set_config(print_changed_only=False)
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict,StratifiedKFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
import pickle
```

```python
df = pd.read_csv('train_traffic.csv')
```

```python
df.head()
```

```python
#divide columns
Y=df['traffic_volume']
X = df.drop(['traffic_volume'],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30, random_state=0)
```

```python
#Run Models on Regular Test/Train Split:
df_reg = pd.DataFrame()
#Linear Regression
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
y_pred = regression_model.predict(X_test)
df_reg.loc[0,'Name'] = 'Liner Regression'
df_reg.loc[0,'Train Score'] = regression_model.score(X_train, y_train)
df_reg.loc[0,'Test Score'] = regression_model.score(X_test, y_test)
df_reg.loc[0,'Mean Squared Error'] = mean_squared_error(y_pred,y_test)
df_reg.loc[0,'MAE'] = mean_absolute_error(y_pred, y_test)
#Random Forest
random_model = RandomForestRegressor(n_estimators=200, random_state=7)
random_model.fit(X_train,y_train)
y_pred = random_model.predict(X_test)
df_reg.loc[1,'Name'] = 'Random Forest'
df_reg.loc[1,'Train Score'] = random_model.score(X_train, y_train)
df_reg.loc[1,'Test Score'] = random_model.score(X_test, y_test)
df_reg.loc[1,'Mean Squared Error'] = mean_squared_error(y_pred,y_test)
df_reg.loc[1,'MAE'] = mean_absolute_error(y_pred, y_test)
#Bagging
bagging_model = BaggingRegressor(n_estimators=200, max_samples= .7, bootstrap=True, oob_score=True, random_state=7)
bagging_model.fit(X_train, y_train)
y_pred = bagging_model.predict(X_test)
df_reg.loc[2,'Name'] = 'Bagging'
df_reg.loc[2,'Train Score'] = bagging_model.score(X_train, y_train)
df_reg.loc[2,'Test Score'] = bagging_model.score(X_test, y_test)
df_reg.loc[2,'Mean Squared Error'] = mean_squared_error(y_pred,y_test)
df_reg.loc[2,'MAE'] = mean_absolute_error(y_pred, y_test)
#Boosting
boosting_model = GradientBoostingRegressor(n_estimators = 200, learning_rate = 0.1, random_state=7)
boosting_model.fit(X_train, y_train)
y_pred = boosting_model.predict(X_test)
df_reg.loc[3,'Name'] = 'Boosting'
df_reg.loc[3,'Train Score'] = boosting_model.score(X_train, y_train)
df_reg.loc[3,'Test Score'] = boosting_model.score(X_test, y_test)
df_reg.loc[3,'Mean Squared Error'] = mean_squared_error(y_pred,y_test)
df_reg.loc[3,'MAE'] = mean_absolute_error(y_pred, y_test)
df_reg
```
